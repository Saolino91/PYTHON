numero di casi totali avvenuti in quello stesso continente

```python
#rlimino NaN dalla colonna continent
continent_cleaned = df_covid.dropna(subset=['continent'])
    
#trovo il valore max per ogni location, dato che total_cases è un valore comulativo
max_cases_per_location = continent_cleaned.groupby(['continent', 'location'])['total_cases'].max()
    
#sommare i massimi per ogni continente per ottenere il totale
total_cases = max_cases_per_location.groupby(level='continent').sum()
    
total_cases.to_frame()
```

Confronto Massimo, media e percentuale rispetto al totale dei contagiati nel mondo risppetto ad Europa ed Asia

```python

import matplotlib.pyplot as plt

# Elimino valori nulli sia per location che per total_cases
cleaned_df = df_covid.dropna(subset=['location', 'total_cases'])

# Filtro Europa e Asia
df_europe = cleaned_df[cleaned_df['continent'] == 'Europe']
df_asia = cleaned_df[cleaned_df['continent'] == 'Asia']

# Trovo il valore massimo di total_cases per ogni location in Europa e li sommo per trovare il totale per continente
max_europe = df_europe.groupby('location')['total_cases'].max().sum()

# Calcola il valore massimo di total_cases per ogni location in Asia e li sommo per trovare il totale per continente
max_asia = df_asia.groupby('location')['total_cases'].max().sum()

# Calcola la media dei total_cases per Europa e Asia
mean_europe = df_europe['total_cases'].mean()
mean_asia = df_asia['total_cases'].mean()

# Calcola il totale dei casi nel mondo
filtered_df= cleaned_df[cleaned_df['location'] == 'World'] # è presente anche total word come location, quindi bisogna utilizzare il massimo di questo valore.
total_world = filtered_df['total_cases'].max()

# Calcola la percentuale dei casi totali per Europa e Asia rispetto al totale mondiale
percentage_europe = (df_europe['new_cases'].sum() / total_world) * 100
percentage_asia = (df_asia['new_cases'].sum() / total_world) * 100

# Risultati in DataFrame
results_europe = pd.DataFrame({
    'Continent': 'Europe',
    'Max Total': [max_europe],
    'Mean': [mean_europe],
    'Percentage': [percentage_europe]
}).round(2)

results_asia = pd.DataFrame({
    'Continent': 'Asia',
    'Max Total': [max_asia],
    'Mean': [mean_asia],
    'Percentage': [percentage_asia]
}).round(2)

# Unisce i risultati in un unico DataFrame
results = pd.concat([results_europe, results_asia]).reset_index(drop=True)

# Visualizzazione Grafica
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# Grafico per il Massimo Totale
results.plot(kind='bar', x='Continent', y='Max Total', ax=axes[0], color='skyblue')
axes[0].set_title('Max Total Cases by Continent')
axes[0].set_ylabel('Total Cases')

# Grafico per la Media
results.plot(kind='bar', x='Continent', y='Mean', ax=axes[1], color='lightgreen')
axes[1].set_title('Mean Total Cases by Continent')
axes[1].set_ylabel('Mean Cases')

# Grafico per la Percentuale
results.plot(kind='bar', x='Continent', y='Percentage', ax=axes[2], color='salmon')
axes[2].set_title('Percentage of New Cases by Continent')
axes[2].set_ylabel('Percentage (%)')

plt.tight_layout()
plt.show()

# Restituisce i DataFrame per revisione
results
```

Grafico evoluzione casi totali covid nel 2022 in Italia

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import calendar 
# Assicurati di operare su una copia per evitare SettingWithCopyWarning
df_italy = df_covid[df_covid['location'] == 'Italy'].copy()
df_italy['date'] = pd.to_datetime(df_italy['date'], errors='coerce')
# Converti la colonna 'date' in datetime usando loc
df_italy.loc[:, 'date'] = pd.to_datetime(df_italy['date'])

# Filtra per l'anno 2022
df_italy_2022 = df_italy[df_italy['date'].dt.year == 2022].copy()

# Aggiungi la colonna 'month' usando loc
df_italy_2022.loc[:, 'month'] = df_italy_2022['date'].dt.month
monthly_cases = df_italy_2022.groupby('month')['total_cases'].sum()

# Crea un grafico a barre
plt.figure(figsize=(10, 6))
plt.bar(monthly_cases.index, monthly_cases.values, color='orange')

# Genera etichette per i mesi
month_labels = [calendar.month_name[i] for i in range(1, 13)]

# Imposta le etichette dei tick per assicurarsi che tutti i mesi siano mostrati
plt.xticks(range(1, 13), month_labels, rotation=45)

plt.title('Nuovi Casi di COVID-19 in Italia per Mese nel 2022')
plt.xlabel('Mese')
plt.ylabel('Nuovi Casi')
plt.tight_layout()
plt.show()


```

Grafico evoluzione nuovi contagiati in italia nel 2022

```python
plt.figure(figsize=(10, 6))
plt.plot(df_italy_2022['date'], df_italy_2022['new_cases'], color='blue', marker='o', linestyle='-', linewidth=1, markersize=5)

#formattazione del grafico
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())

plt.title('Evoluzione dei Nuovi Casi di COVID-19 in Italia nel 2022')
plt.xlabel('Data')
plt.ylabel('Nuovi Casi')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
plt.show()
```

confronto frafico somma cumulativa nuovi casi e grafico dei casi totali, si nota che hanno la stessa traiettoria

```python
#calcolo la somma cumulativa dei nuovi casi
df_italy_2022['cumulative_new_cases'] = df_italy_2022['new_cases'].cumsum()

#grafico dell'andamento cumulativo dei nuovi casi
plt.figure(figsize=(12, 8))
plt.plot(df_italy_2022['date'], df_italy_2022['cumulative_new_cases'], label='Somma Cumulativa Nuovi Casi', color='green', linestyle='--')
plt.plot(df_italy_2022['date'], df_italy_2022['total_cases'], label='Casi Totali', color='blue', linestyle='-')

#formattazione del grafico
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%B'))
plt.gca().xaxis.set_major_locator(mdates.MonthLocator())
plt.xticks(rotation=45)
plt.title('Andamento Somma Cumulativa Nuovi Casi vs Casi Totali in Italia nel 2022')
plt.xlabel('Data')
plt.ylabel('Numero di Casi')
plt.legend()
plt.tight_layout()
plt.show()
```

Boxplot che confronta i pazienti in terapia intensiva Italia, Germania e France tra maggio 2022 e aprile 2023. Dal grafico si evince: Francia: mostra alcuni outlier sulla parte bassa, indicando alcuni giorni in cui il numero di pazienti in terapia intensiva era eccezionalmente basso.
Germania: ampia distribuzione dei dati, con una mediana più alta della Francia, e la presenza di outlier sia in alto che in basso, indicando che ci sono stati sia giorni con un numero molto alto sia giorni con un numero molto basso di pazienti in terapia intensiva.
Italia: Lmostra il box più stretto, il che indica una variazione minore rispetto agli altri due paesi e nessun outlier evidente. Questo suggerisce che il numero di pazienti in terapia intensiva in Italia era relativamente stabile durante il periodo considerato.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# filtro nazioni interessate
df_countries = df_covid[df_covid['location'].isin(['Italy', 'Germany', 'France'])].copy()  # Usa .copy() per evitare warning

# converto la colonna Date in formato data con .loc[]
df_countries.loc[:, 'date'] = pd.to_datetime(df_countries['date'])

# filtro il periodo da maggio 2022 ad aprile 2023
start_date = pd.to_datetime('2022-05-01')
end_date = pd.to_datetime('2023-04-30')
df_period = df_countries[(df_countries['date'] >= start_date) & (df_countries['date'] <= end_date)]

# creazione del boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x='location', y='icu_patients', data=df_period)
plt.title('Pazienti in Terapia Intensiva (Maggio 2022 - Aprile 2023)')
plt.xlabel('Nazione')
plt.ylabel('Numero di Pazienti in Terapia Intensiva')
plt.show()
```
