```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(override=True)
username=os.getenv("username")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)
dimcustomer = """SELECT * FROM dimcustomer
        """
dimcustomer = pd.read_sql(dimcustomer, db_engine)
db_engine.dispose()
dimcustomer.shape
dimcustomer
```

```python
dimcustomer["FirstName"] = dimcustomer["FirstName"].str.lower().to_frame()
dimcustomer["LastName"] = dimcustomer["LastName"].str.upper().to_frame()
dimcustomer.head()
```

```python
dimcustomer[['Utente', 'Dominio']] = dimcustomer['EmailAddress'].str.split("@", expand=True)
dimcustomer[['Utente', 'Dominio']]
```

```python
pattern = r'(\d+) (\(\d+\)) (\d+) (\d+-\d+)'
phone_parts = dimcustomer['Phone'].str.extract(pattern)

phone_parts.columns = ['Paese', 'Area', 'Prefisso', 'Numero']
phone_parts
```

```python
filtro_email2 = dimcustomer[dimcustomer['EmailAddress'].str.contains('10|20')]
filtro_email2[['EmailAddress']]
```

```python
dimcustomer['EmailLength'] = dimcustomer['EmailAddress'].str.len()

dimcustomer_ordinati = dimcustomer.sort_values(by='EmailLength', ascending=False)

top_10_lunghi = dimcustomer_ordinati.head(10)

top_10_corti = dimcustomer_ordinati.dropna(subset=['EmailAddress']).tail(10)

combined_results = pd.concat([top_10_lunghi, top_10_corti], keys=['Longest', 'Shortest'])

combined_results[['EmailAddress', 'EmailLength']]
```

```python
import os
import pandas as pd

# Ottieni l'elenco dei file nella directory 'beginner_datasets/'
files = os.listdir("beginner_datasets/")

# Lista per tenere traccia dei file con valori nulli
files_con_nulli = []

# 1. Itera su ogni file nella directory
for file in files:
    # 2. Seleziona solo i file con estensione .csv
    if file.endswith('.csv'):
        # Costruisci il percorso completo del file
        file_path = os.path.join("beginner_datasets/", file)
        
        # 3. Leggi il file e caricalo in un DataFrame
        try:
            df = pd.read_csv(file_path)
            
            # 4. Utilizza .isna() per trovare la maschera booleana dei dati nulli
            # 5. Conta i dati nulli utilizzando .sum(); utilizzalo due volte per ottenere il totale dei valori nulli
            num_nulli = df.isna().sum().sum()  # Il primo .sum() somma per colonna, il secondo somma queste somme
            
            # 6. Se il DataFrame contiene valori nulli, stampa o memorizza il nome del file
            if num_nulli > 0:
                files_con_nulli.append(file)
                
        except Exception as e:
            print(f"Non è stato possibile elaborare il file {file}: {e}")

# Stampa l'elenco dei file che contengono valori nulli
print("Files con valori nulli:", files_con_nulli)
```
