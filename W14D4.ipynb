```python
dfp = pd.DataFrame(pd.read_csv("beginner_datasets/pokemon.csv")) #dataset pokemon
```

```python
dfp.set_index("#") # collocata la colonna # come index
```

```python
legendary_pokemon = dfp[dfp["Legendary"] == True]["Name"]
print(legendary_pokemon) #pokemon leggendari
```

```python
legendary_pokemon_grass = dfp[(dfp["Type 1"] == "Grass") & (dfp["Legendary"] == True)]["Name"]
print(legendary_pokemon_grass) #leggendari di tipo grass
```

```python
legendary_pokemon_fire_ice = dfp[((dfp["Type 1"] == "Fire") | (dfp["Type 1"] == "Ice")) & (dfp["Legendary"] == True)]["Name"]
print(legendary_pokemon_fire_ice) #pokemon leggendari di tipo fuoco o ghiaccio
```

```python
dfp.sort_values("Name") # ordinare in base al nome crescente e colonna name come Index
dfp.set_index("Name")
```

```python
import dotenv
import os
from sqlalchemy import create_engine
import pandas as pd
dotenv.load_dotenv(override=True)
username=os.getenv("username")
password=os.getenv("password")
host=os.getenv("host")
dbname=os.getenv("dbname")
conn_string = f"mysql+pymysql://{username}:{password}@{host}/{dbname}"
db_engine = create_engine(conn_string)
query = """SELECT * FROM dimproduct
        """
df = pd.read_sql(query, db_engine)
db_engine.dispose()
df.shape
df
```

```python
avg_price = df.groupby("EnglishProductName")["DealerPrice"].mean()
print(avg_price) #costo medio prodotto
```

```python
percentile75 = df['ListPrice'].quantile(0.75) #range quartile più costoso

maggiore_percentile75 = df[df['ListPrice'] >= percentile75]

min = maggiore_percentile75['ListPrice'].min()
max = maggiore_percentile75['ListPrice'].max()

percentile75, min, max
```

```python
color_price_avg = df.groupby("Color")["DealerPrice"].mean() # prodotti di colore rosso blu e nero
blu = color_price_avg.loc[["Blue"]]
red_black = color_price_avg.loc[["Red", "Black"]]

print(blu)
print(red_black)
```
